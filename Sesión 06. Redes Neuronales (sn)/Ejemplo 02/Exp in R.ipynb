# Ahora usemos R para generar un an√°lisis exploratorio de datos

```r
library(ggplot2)
library(dplyr)
```

# Hagamos un histograma con ggplot2 de las edades

```r
data = read.csv("pima-indians-diabetes-for-R.csv")

data %>% ggplot(aes(x= edad)) +
  geom_histogram() +
  geom_vline(aes(xintercept=mean(data$edad)),
            color="blue", linetype="dashed", size=1) +
  geom_histogram(color="black", fill="white") +
  geom_histogram(aes(y=..density..), colour="black", fill="white")+
  geom_density(alpha=.2, fill="#FF6666") 
```

# Exploremos los casos observados de diabetes

```r
data %>% ggplot(aes(x= dummy)) +
  geom_bar()
```
